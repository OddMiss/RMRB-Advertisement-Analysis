PM Fix: To know if your factor actually possesses Alpha, you should compare its returns against an Equal-Weighted Index of all industries, not a cap-weighted mega-cap index.

Instead of using the media Z-score to pick the stocks, use a classic, proven financial factor (like 20-day Price Momentum or Value/Low PE) to select the top industries. Then, use your Media Z-score as a gatekeeper:

- Rule: Only buy a high-momentum industry if its media attention is also low (Stealth Institutional Accumulation).

- Rule: Instantly sell a high-momentum industry if its media attention spikes to an extreme high (Retail Overcrowding/The Bubble Bursting).

In institutional quantitative finance, keeping non-trading days (filled with 0.0 returns) in your final analytics suite creates a form of "statistical pollution." Removing them and using a pure trading-day index is the industry standard because it corrects significant distortions in risk calculations.
